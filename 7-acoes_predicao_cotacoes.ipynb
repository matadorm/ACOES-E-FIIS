```python
import pandas as pd
import numpy as np
```

```python
df_cotacoes_justas = pd.read_csv("C:\\Users\\Matheus Poletto\\Desktop\\Cientista de Dados\\ACOES-E-FIIS\\cotacoes_justas.csv")
df_cotacoes_atuais = pd.read_csv("C:\\Users\\Matheus Poletto\\Desktop\\Cientista de Dados\\ACOES-E-FIIS\\stocks_rvi_fund_07-01-2025.csv")
```

```python
df_cotacoes_justas = df_cotacoes_justas.rename(columns={'Cotacao': 'Cotacao_justa'})
df_cotacoes_justas
```

```python
#df_cotacoes_atuais = df_cotacoes_atuais[['ticker', 'Cotacao']]
df_cotacoes_atuais = df_cotacoes_atuais.rename(columns={'Cotacao': 'Cotacao_atual'})

df_cotacoes_atuais
```

```python
# Merge das tabelas no ticker
df_completo = pd.merge(df_cotacoes_justas, df_cotacoes_atuais, on='ticker')

# CriaÃ§Ã£o da variÃ¡vel de valorizaÃ§Ã£o
df_completo["Potencial Upside/Downside (%)"] = (1-(df_completo['Cotacao_atual']/df_completo['Cotacao_justa']))*100

# Obter a coluna 'Potencial Upside/Downside (%)'
coluna = df_completo.pop('Potencial Upside/Downside (%)')

# Inserir a coluna na terceira posiÃ§Ã£o (Ã­ndice 2)
df_completo.insert(2, 'Potencial Upside/Downside (%)', coluna)

# Ordenando conforme variÃ¡vel criada
df_completo = df_completo.sort_values(by='Potencial Upside/Downside (%)', ascending=False)

#df_completo[df_completo['ticker'] == 'PRIO3']
```
