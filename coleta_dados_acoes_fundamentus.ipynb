```python
papeis = {
    'Papel':[],
    'Cotacao':[],
    'P/L':[],
    'P/VP':[],
    'PSR':[],
    'DY':[],
    'P/Ativo':[],
    'P/Cap.Giro':[],
    'P/EBIT':[],
    'P/ACL':[],
    'EV/EBIT':[],
    'EV/EBITDA':[],
    'Mrg.Ebit':[],
    'Mrg.Liq.':[],
    'Liq.Corr.':[],
    'ROIC':[],
    'ROE':[],
    'Liq.2meses':[],
    'Pat.Liq':[],
    'Div.Brut/Pat.':[],
    'Cresc.5anos':[]}

for papel, info in data.items():
    papeis['Papel'].append(papel)
    papeis['Cotacao'].append(info['Cotacao'])
    papeis['P/L'].append(info['P/L'])
    papeis['P/VP'].append(info['P/VP'])
    papeis['PSR'].append(info['PSR'])
    papeis['DY'].append(info['DY'])
    papeis['P/Ativo'].append(info['P/Ativo'])
    papeis['P/Cap.Giro'].append(info['P/Cap.Giro'])
    papeis['P/EBIT'].append(info['P/EBIT'])
    papeis['P/ACL'].append(info['P/ACL'])
    papeis['EV/EBIT'].append(info['EV/EBIT'])
    papeis['EV/EBITDA'].append(info['EV/EBITDA'])
    papeis['Mrg.Ebit'].append(info['Mrg.Ebit'])
    papeis['Mrg.Liq.'].append(info['Mrg.Liq.'])
    papeis['Liq.Corr.'].append(info['Liq.Corr.'])
    papeis['ROIC'].append(info['ROIC'])
    papeis['ROE'].append(info['ROE'])
    papeis['Liq.2meses'].append(info['Liq.2meses'])
    papeis['Pat.Liq'].append(info['Pat.Liq'])
    papeis['Div.Brut/Pat.'].append(info['Div.Brut/Pat.'])
    papeis['Cresc.5anos'].append(info['Cresc.5anos'])


df_acoes = pd.DataFrame(papeis)
df_acoes
```

# Tratamento dos dados

## Removendo valores 0.00 do PL e Liq.2meses

```python
df_acoes['P/L'] = df_acoes['P/L'].replace(0.00, '')
df_acoes['P/L'] = pd.to_numeric(df_acoes['P/L'], errors='coerce')
df_acoes['Liq.2meses'] = pd.to_numeric(df_acoes['P/L'], errors='coerce')
df_acoes
```

### Após método de conversão testado, transformação em função

```python
def format_to_numeric_column (df, col_name):
    df[col_name] = pd.to_numeric(df[col_name])

def format_to_percent_column (df,col_name):
    df[col_name] = df[col_name].apply(lambda x: f'{x:.2f}%')

columns_format = ['DY', 'Mrg.Ebit','Mrg.Liq.','ROE','ROIC', 'Cresc.5anos']

for col in columns_format:
    format_to_numeric_column(df_acoes, col)

for col in columns_format:
    format_to_percent_column(df_acoes, col)

df_acoes
```
