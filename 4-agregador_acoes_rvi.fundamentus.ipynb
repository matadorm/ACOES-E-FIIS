```python
import pandas as pd
```

# Junção das tabelas RVI e Fundamentus

```python
df_fund = pd.read_csv("data_fund\\stocks_fund_07-01-2025.csv")
df_rvi = pd.read_csv("data_yfinance\\stocks_rvi_07-01-2025.csv")
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df_rvi.head(10)
```

```python
df_c = pd.merge(df_fund, df_rvi, on='ticker')
df_c.head(10)
```

```python
df_c['LPA'] = (df_c['Cotacao']/df_c['P/L'])
df_c.head(10)
```

```python
columns_drop = ['rvi_Conservador', 'rvi_Moderado', 'rvi_Arrojado']
df_c.drop(columns=columns_drop, inplace=True)
df_c.head(10)
```

```python
columns = df_c.columns.tolist()
columns.insert(17, columns.pop(columns.index('LPA')))
df_c = df_c[columns]
pd.set_option('display.max_rows', 10)
df_c
```

```python
df_c = df_c.loc[df_c['P/L'] > 0]
df_c
```

```python
df_c.sort_values(by='avg_return',ascending=False, inplace=True)
df_c.head(10)
```

```python
df_c.to_csv('stocks_rvi_fund_07-01-2025.csv', index=False)
```
